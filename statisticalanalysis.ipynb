This file contains the progress of our statistical analysis. 

```python
import pandas as pd
```

```python
dataframe = pd.DataFrame.from_csv('h1b_kaggle.csv')
cleandf = dataframe[dataframe.lat.notnull()]
cleandf = cleandf[cleandf.lon.notnull()]
cleandf = cleandf[cleandf.SOC_NAME.notnull()]
dataframe = cleandf
dataframe
```

Job title and case status statistics

```python
#Extract top 50 jobs in dataset
newdf = pd.DataFrame()
newdf['Counts'] = dataframe['SOC_NAME'].value_counts()
newdf['SOC_NAME'] = newdf.index
newdf.sort_values(by = ['Counts'], ascending = True)
topjobs = newdf.head(50)['SOC_NAME'].tolist()
topjobs
```

```python
jobtotalapps = []
jobcertified = []
for job in topjobs:
    thisjobdf = dataframe[(dataframe['SOC_NAME'] == job)]
    jobtotalapps.append(len(thisjobdf))
    thisjobdfc = thisjobdf[(thisjobdf['CASE_STATUS']).str.contains("CERTIFIED")]
    jobcertified.append(len(thisjobdfc))

print(jobtotalapps)
print(jobcertified)

    
```

```python
percentcert = []
for i in range(len(topjobs)):
    percentcert.append(jobcertified[i] / jobtotalapps[i])

print(percentcert)
```

```python
jobcertdf = pd.DataFrame()
jobcertdf['SOC_NAME'] = topjobs
jobcertdf['Percent_Certified'] = percentcert
jobcertdf
```

Top employers with most applicants statistics

```python
#Extract top 50 employers in dataset
employersdf = pd.DataFrame()
employersdf['Counts'] = dataframe['EMPLOYER_NAME'].value_counts()
employersdf['EMPLOYER_NAME'] = employersdf.index
employersdf.sort_values(by = ['Counts'], ascending = True)
topemployers = employersdf.head(50)['EMPLOYER_NAME'].tolist()
topemployers
```

```python
employertotalapps = []
employercertified = []
for employer in topjobs:
    thisjobdf = dataframe[(dataframe['EMPLOYER_NAME'] == employer)]
    employertotalapps.append(len(thisjobdf))
    thisjobdfc = thisjobdf[(thisjobdf['CASE_STATUS']).str.contains("CERTIFIED")]
    employercertified.append(len(thisjobdfc))

print(employertotalapps)
print(employercertified)
```

```python
percentcertemp = []
for i in range(len(topemployers)):
    percentcertemp.append(employercertified[i] / employertotalapps[i])
print(percentcertemp)
```

```python
employercertdf = pd.DataFrame()
employercertdf['EMPLOYER_NAME'] = topemployers
employercertdf['Percent_Certified'] = percentcertemp
employercertdf
```
